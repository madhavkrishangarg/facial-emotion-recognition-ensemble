```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import load_img
import tensorflow as tf
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import Callback, EarlyStopping, ReduceLROnPlateau

from PIL import Image
from keras.utils import np_utils
import scikitplot


#These are the paths for the directories, where we are importing the dataset
train_dir = "../input/fer2013/train/" 
test_dir = "../input/fer2013/test/"  
```

```python
#This is the function in order to calculate the expectation
def calculate_expectation(path, index_values):
    #Dictionary
    dictionary_info = {}
    
    #These are the images of different emotions
    for expression in os.listdir(path):
        directory_path = path + expression
        dictionary_info[expression] = len(os.listdir(directory_path))
    #Converting to dataframe and then returning
    df = pd.DataFrame(dictionary_info, index=[index_values])
    return df

#Calculating total amount of values
no_training = calculate_expectation(train_dir, 'train')
test_count = calculate_expectation(test_dir, 'test')
print(no_training)
print(test_count)
no_training.transpose().plot(kind='bar')
```

```python
#This is for the testing data plotting 
test_count.transpose().plot(kind='bar')
```

```python
plt.figure(figsize=(28, 44))

#Initializing the loop variable
#Depending on what will be the position of the printed out image
i = 1

#Running the for loop to iterate over the whole training directory
for expression in os.listdir(train_dir):
    #This is to print out random 5 images belonging to each of the class
    for j in range(0, 5):
        #Image loading
        img = load_img((train_dir + expression +'/'+ os.listdir(train_dir + expression)[random.randint(0,111)]))
        plt.subplot(7, 5, i)
        #Printing out the image
        plt.imshow(img)
        plt.title(expression)
        plt.axis('off')
        i += 1
        
#Showing the whole plot
plt.show()
```

Now we are going to check for the different sizes which are present in the images
We are maintaining a dictionary and then we are adding all the images to it

```python
#To maintain, how many images belonging to which classes
size_dict = {}

#Running for loop over the training dataset
for expression in os.listdir(train_dir):
    expression_dir = os.path.join(train_dir, expression)
    #Going through all the expressions
    for filename in os.listdir(expression_dir):
        #Path of the image
        img_path = os.path.join(expression_dir, filename)
        img = Image.open(img_path)
        img_size = img.size
        #Adding value to the images
        if img_size in size_dict:
            size_dict[img_size] += 1
        else:
            size_dict[img_size] = 1

print(size_dict)
```
