```python
import os
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import numpy as np


import matplotlib.pyplot as plt


def plot_map(history):
    plt.plot( list(range(1, 1+len(history.history['loss']))), history.history['loss'])
    plt.plot(list(range(1, 1+len(history.history['val_loss']))), history.history['val_loss'])
    plt.xlabel("Epochs")
    plt.legend(['Training loss', 'Validation loss'])
    plt.show()
    
    plt.plot( list(range(1, 1+len(history.history['accuracy']))), history.history['accuracy'])
    plt.plot(list(range(1, 1+len(history.history['val_accuracy']))), history.history['val_accuracy'])
    plt.xlabel("Epochs")
    plt.legend(['Training accuracy', 'Validation accuracy'])
    plt.show()
```

```python
#This cell is for data importing, will be used for all the models
path_val = ("data_folder\\images\\images\\train")
emotions = os.listdir("data_folder\\images\\images\\train")

#Let's bulid a 2 dimensional array containing all the names of the images for each of the emotions
images = [[]]*len(emotions)

for i in range(len(emotions)):
    images[i] = os.listdir(os.path.join(path_val, emotions[i]))


from tqdm import notebook

angry = []
disgust = []
fear = []
happy = []
neutral = []
sad = []
surprise = []
for i in notebook.tqdm(range(len(images[0]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[0], images[0][i])))
    angry += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

for i in notebook.tqdm(range(len(images[1]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[1], images[1][i])))
    disgust += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

for i in notebook.tqdm(range(len(images[2]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[2], images[2][i])))
    fear += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

for i in notebook.tqdm(range(len(images[3]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[3], images[3][i])))
    happy += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

for i in notebook.tqdm(range(len(images[4]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[4], images[4][i])))
    neutral += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

for i in notebook.tqdm(range(len(images[5]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[5], images[5][i])))
    sad += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

for i in notebook.tqdm(range(len(images[6]))):
    image_gray = np.array(Image.open(os.path.join(path_val, emotions[6], images[6][i])))
    surprise += [np.repeat(image_gray[:, :, np.newaxis], 3, axis=2)]

images_array = [angry, disgust, fear, happy, neutral, sad, surprise]

#Building numpy arrays for training and testing from the orifinal values
y = []
x = []
for i in (range(len(images_array))):
    for j in (range(len(images_array[i]))):
        y += [i]
        x += [images_array[i][j]]
        

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
       
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rotation_range=5,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

#We are upscalling the data to 8000 images each
angry_val_aug = []
disgust_val_aug = []
fear_val_aug = []
happy_val_aug = [] 
neutral_val_aug = []
sad_val_aug = []
surprise_val_aug = []

aug_imgs = [angry_val_aug, disgust_val_aug, fear_val_aug, happy_val_aug, neutral_val_aug, sad_val_aug, surprise_val_aug]

for i in notebook.tqdm(range(len(images_array))):
    req_more = 8000 - len(images_array[i])
    for j in notebook.tqdm(range(req_more)):
        choose_img = images_array[i][np.random.choice(len(images_array[i]))]
        aug_img = datagen.random_transform(choose_img)
        aug_imgs[i] += [aug_img]
        
```

```python
x_aug = []
y_aug = []

x_aug += x
y_aug += y

for i in (range(len(aug_imgs))):
    for j in (range(len(aug_imgs[i]))):
        y += [i]
        x += [aug_imgs[i][j]]
```

```python
from sklearn.model_selection import train_test_split

X_train_aug, X_test_aug, y_train_aug, y_test_aug = train_test_split(x_aug, y_aug, test_size=0.3, random_state=42)
       
```

```python
base_model_4 = tf.keras.applications.ResNet50(input_shape=(48,48,3),include_top=False,weights="imagenet")

for i in base_model_4.layers[:-5]:
    i.trainable = False
    
output_model = base_model_4.output
output_model = Flatten()(output_model)
output_model = BatchNormalization()(output_model)
output_model = Dense(128 , activation = "relu")(output_model)
output_model = BatchNormalization()(output_model)
output_model = Dropout(0.3)(output_model)
output_model = Dense(64 , activation = "relu")(output_model)
output_model = BatchNormalization()(output_model)
output_model = Dropout(0.3)(output_model)
output_model = Dense(32 , activation = "relu")(output_model)
output_model = BatchNormalization()(output_model)
output_model = Dropout(0.3)(output_model)
output_model = Dense(7 , activation  = "softmax")(output_model)

model_ite6 = Model(inputs = base_model_4.input, outputs = output_model)

model_ite6.compile(tf.keras.optimizers.Adam(), loss='SparseCategoricalCrossentropy', metrics = ['accuracy'])

history_iter6 = model_ite6.fit(np.array(X_train_aug)/255., np.array(y_train_aug), epochs = 10, validation_data = (np.array(X_test_aug)/255., np.array(y_test_aug)), workers=6, use_multiprocessing=True)
plot_map(history_iter6)
```

Data augmentation-part2

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   validation_split = 0.2,
                                  
        rotation_range=5,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest')

valid_datagen = ImageDataGenerator(rescale = 1./255,validation_split = 0.2)

test_datagen  = ImageDataGenerator(rescale = 1./255)
                                  
```

```python
train_dataset  = train_datagen.flow_from_directory(directory = 'data_folder\\images\\images\\train',
                                                   target_size = (48,48),
                                                   class_mode = 'categorical',
                                                   subset = 'training',
                                                   batch_size = 64)
valid_dataset = valid_datagen.flow_from_directory(directory = 'data_folder\\images\\images\\train',
                                                  target_size = (48,48),
                                                  class_mode = 'categorical',
                                                  subset = 'validation',
                                                  batch_size = 64)

test_dataset = test_datagen.flow_from_directory(directory = 'data_folder\\images\\images\\validation',
                                                  target_size = (48,48),
                                                  class_mode = 'categorical',
                                                  batch_size = 64)
```

```python
test_model = tf.keras.applications.ResNet50(input_shape=(48,48,3),include_top=False,weights="imagenet")

for layer in test_model.layers[:-4]:
    layer.trainable=False

model=Sequential()
model.add(test_model)
model.add(Dropout(0.5))
model.add(Flatten())
model.add(BatchNormalization())
model.add(Dense(32,kernel_initializer='he_uniform'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(32,kernel_initializer='he_uniform'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(32,kernel_initializer='he_uniform'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(7,activation='softmax'))

model.compile(tf.keras.optimizers.Adam(), loss='categorical_crossentropy', metrics = ['accuracy'])

history_iter7 = model.fit(train_dataset,validation_data=valid_dataset, epochs = 10)
plot_map(history_iter7)
```
